```python
import matplotlib.pyplot as plt
import pickle
with open('files/train_losses.pickle', 'rb') as handle:
    train_losses = pickle.load(handle)
with open('files/val_losses.pickle', 'rb') as handle:
    val_losses = pickle.load(handle)
with open('files/train_accs.pickle', 'rb') as handle:
    train_accs = pickle.load(handle)
with open('files/val_accs.pickle', 'rb') as handle:
    val_accs = pickle.load(handle)
with open('files/learning_rates.pickle', 'rb') as handle:
    learning_rates = pickle.load(handle)

print(len(train_losses))
print(len(val_losses))
print(len(train_accs))
print(len(val_accs))
print(len(learning_rates))
```

```python
import matplotlib.pyplot as plt


# Create the plot
plt.figure(figsize=(10, 6))  # Adjust figure size as desired

# Plot train_losses
plt.plot(train_losses, label='Training Loss')

# Plot val_losses
plt.plot(val_losses, label='Validation Loss')

# Add labels and title
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Comparison of Training and Validation Loss')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


# Create the plot
plt.figure(figsize=(10, 6))  # Adjust figure size as desired

# Plot train_losses
plt.plot(train_accs, label='Training Accuracy')

# Plot val_losses
plt.plot(val_accs, label='Validation Accuracy')

# Add labels and title
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Comparison of Training and Validation Accuracy')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
